```python
def count_pixel_percent(columns_properties):
    
    for col_prop in columns_properties:
        col_range = col_prop['c_range']
        d = max(int(1000 / col_range), 1)
        
        col_prop['col_width'] = d * col_range
        
        for area in col_prop['areas']:

            area['c_position'] = int(area['c_min']) * d            
            area['c_width'] = (int(area['c_max']) - int(area['c_min'])) * d
            
            
def make_properties(not_explored_chunks, catalogue_item):
    
    columns_properties = []
    columns_ranges = []
    # chunks columns have to be in same order
    # catalogue item has to have colums in same order as chunks
    for i, col in enumerate(not_explored_chunks[0]['borders']):
        col_prop = {}
        
        # in catalogue item value min - first, value max - last
        # ci distribution mustn't be short
        ci_col_dist_min = catalogue_item['spec'][i]['distribution'][0]['value_min']
        ci_col_dist_max = catalogue_item['spec'][i]['distribution'][-1]['value_max']
        ci_col_range = ci_col_dist_max-ci_col_dist_min
        
        col_prop['c_name'] = str(col['column'])
        col_prop['c_type'] = str(col['type'])
        col_prop['areas'] = []
        col_prop['c_range'] = ci_col_range
        
        for ch in not_explored_chunks:
            
            ch_min = ch['borders'][i]['minimum']
            ch_max = ch['borders'][i]['maximum']
            ch_range = abs(ch_max-ch_min)
                
            width = (ch_range/ci_col_range)
            position = (ch_min/ci_col_range)
            width = round(width, 2)
            position = round(position, 2)
            
            col_prop['areas'].append({
                'c_min': ch_min,
                'c_max': ch_max,
                'c_position': position,
                'c_width': width,
            })
            
        col_prop['areas'] = sorted(col_prop['areas'], key=lambda x: x['c_min'])
        for area in col_prop['areas']:
            area['c_min'] = str(area['c_min'])
            area['c_max'] = str(area['c_max'])
            
        columns_properties.append(col_prop)

    count_pixel_percent(columns_properties)
    
    return columns_properties

maked_properties = make_properties(chunks, ci)
maked_properties
```

```python
def make_area(c_min, c_max, c_position, c_width):

    return f'''
    <div 
        class="area" 
        style="left: {c_position}px; width: {c_width}px;">        
    </div>  
    <div
        class="area-label"     
        style="left: {c_position}px; top: 50px; width: 10px;">{c_min}</div>
    <div
        class="area-label"     
        style="left: {c_position + c_width}px; top: 50px; width: 10px;">{c_max}</div>        
    '''

def make_areas(areas):
    return ''.join([make_area(**area) for area in areas])


def make_column(c_name, c_type, areas, col_width, **kwargs):

    return f'''
    <h3>
        {c_name} ({c_type})
    </h3>
    <div class="column" style="width: {col_width}px">
        {make_areas(areas)}
    </div>
    '''
    
def make_columns(columns_properties):    
    return ''.join([make_column(**col_prop) for col_prop in columns_properties])   

display(HTML(f'''
    <html>
    <head>
    <style>
    .column {{
        position: relative;
        margin-top: 5px;
        height: 50px;       
        background: #f2f2f2;
    }}
    .area {{
        position: absolute;
        background: #ff8080;
        height: 50px;
    }}    
    .area-label {{
        position: absolute;
    }} 
    h2, h3 {{
        color: #333333;
    }}
    </style>
    </head>
    <body>
    
        <h2>NOT EXPLORED CHUNKS</h2>       
        {make_columns(maked_properties)}

    </body>
    </html> 
'''))
```
